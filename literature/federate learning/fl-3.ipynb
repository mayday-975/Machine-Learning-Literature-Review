# How to Share a Secret 
- 1979
- Adi Shamir
- Massachusetts Institute of Technology

# Abstract

- 提出一种方法((k, n) threshold scheme, (k,n)门限秘密分享方案)，将数据D分成n个块儿，并且能从n块中取k个块，用它们来恢复出数据D
- 在加密系统中，当出现一些意外的破坏时，此方法能安全且可靠地利用部分数据来恢复整体



# Introduction

- 目标，将D分成片:$D_1, D_2, \dots , D_n$，并满足
    - 任意k或k+个$D_i$可以求出D
    - 任意k-1或更少的$D_i$，完全求不出D
- 过去工作 
    - 可通过对数据进行加密的方法来保护数据，但是保护加密的密钥则需要用不同的方法
    - 最安全的密钥管理方法是将密钥放到一个单一的，受到保护的地方(一台电脑，人脑，或者密码箱)
        - 问题：非常不可靠，如果发生意外(电脑坏了或者人去世了，密码箱被破坏)
    - 另一种方法：将密钥拷贝多份保存到多个地方
        - 问题：增加安全风险(电脑被入侵，人为错误，背叛等)
    - 本方案
    - (k,n) threshold scheme with n = 2k -1. 即使没了k-1(比如被偷)个片依然能恢复，且敌人有了k-1片，也无法恢复出原始密钥
    
- 两种tradeoff
     - 安全和可靠，上文介绍的内容
     - 安全和易用
         - 想象一个场景，一个文件需要公司多个高层签名，但是原来的方法是让他们都去签名，这就很不方便。但如果采用(3,n) threshold scheme。就只需要其中3个人签名，就可以通过了，增加公司的办事效率。




# A Simple (k, n) Threshold Scheme

- 方法基于多元插值法
    - 想象有k个点在一个二维平面上$(x_i,y_i)...(x_k,y_k)$
    - 有且仅有一个k-1次多项式函数$q(x)$满足$q(x_i) = y_i$
    - 把D想象成一个数字
    - 随机假设一个k-1次多项式函数$q(x) = a_0+a_1x+...a_{k-1}x^{k-1}$，满足：
        - $a_0=D$
        - $D_1=q(1), ... , D_i=q(i), ... , D_n = q(n)$
    - 因此，通过任意k个$(i,D_i)$对，就能解出$q(x)$，随后得到$D = q(0)$



- 优点
    - 每个数据片的大小不会超过原数据
    - 当k固定时，某一个数据片$D_i$可以被动态地添加或删除
        - 例如，公司中，有高层离开或加入这个公司
    - 改变所有的数据片$D_i$只需要找一个新的$q(x)$，不用动原数据。此外，较为频繁更换$D_i$有助于提高安全性
    - 可以用来进行重要性/层级分配。
        - 例如，一个公司里，董事长可以分三个数据片$D_i$，CEO两个，中层管理一个...等
        
- 问题
    - D的数据类型

- 拉格朗日插值法
    - 已知k个数据点

$$
\left\{\left(x_{i_{1}}, s_{i_{1}}\right), \cdots,\left(x_{k_{t}}, s_{k_{t}}\right)\right\}
$$
        
- 求 $g(x)$

$$
g(x)=\sum_{r=1}^{k} S_{i_{r}} \prod_{t=1 \atop t \neq r}^{k} \frac{x-x_{i_{i}}}{x_{i_{r}}-x_{i_{t}}}
$$
